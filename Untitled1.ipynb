```python
import json
import torch
from unsloth import FastLanguageModel
from datasets import Dataset
from trl import SFTTrainer
from transformers import TrainingArguments

print(f"GPU: {torch.cuda.get_device_name(0)}")
```

```python
MODEL_NAME = "unsloth/Qwen2.5-Coder-7B-Instruct-bnb-4bit"
MAX_SEQ_LENGTH = 768

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name=MODEL_NAME,
    max_seq_length=MAX_SEQ_LENGTH,
    dtype=None,  # auto-detect
    load_in_4bit=True,
)

print("Model loaded!")
```

```python
model = FastLanguageModel.get_peft_model(
    model,
    r=16,
    lora_alpha=32,
    target_modules=["q_proj", "k_proj", "v_proj", "o_proj",
                     "gate_proj", "up_proj", "down_proj"],
    lora_dropout=0.05,
    bias="none",
    use_gradient_checkpointing="unsloth",
    random_state=42,
)

trainable = sum(p.numel() for p in model.parameters() if p.requires_grad)
total = sum(p.numel() for p in model.parameters())
print(f"Trainable: {trainable:,} / {total:,} ({100*trainable/total:.2f}%)")
```

```python
import os
import glob
import json
from datasets import Dataset

# 1. Load all JSON batches and apply Complexity Weighting
raw_data = []
batch_files = glob.glob("dataset_batches/*.json")

if not batch_files:
    print("Warning: No JSON files found in the 'dataset_batches' folder! Please check your upload.")

for file in batch_files:
    with open(file, "r", encoding="utf-8") as f:
        batch = json.load(f)
        for item in batch:
            # Multiply the example by its weight (e.g., DP = 3 copies, Sorting = 1 copy)
            weight = int(item.get("complexity_weight", 1.0))
            raw_data.extend([item] * weight)

print(f"Loaded {len(batch_files)} batches.")
print(f"Total training examples (after complexity weighting): {len(raw_data)}")

# 2. Format the prompt using the new keys
def format_example(example):
    # Dynamically inject the language (C++ or Python) to give the model better context
    lang = "C++" if example['language'] == "cpp" else "Python"

    text = (
        "<|im_start|>system\n"
        "You are a code commenting assistant. Add short inline comments to the end of each line of code. "
        "Use // for C++ and # for Python. Keep comments under 10 words. "
        "Return ONLY the commented code.<|im_end|>\n"
        "<|im_start|>user\n"
        f"Add inline comments to this {lang} code:\n\n{example['code_without_comments']}<|im_end|>\n"
        "<|im_start|>assistant\n"
        f"{example['code_with_comments']}<|im_end|>"
    )
    return {"text": text}

dataset = Dataset.from_list(raw_data)
dataset = dataset.map(format_example)

# 3. Filter out examples that exceed your context window limit
def filter_long(example):
    # Assuming `tokenizer` and `MAX_SEQ_LENGTH` are defined in an earlier cell
    tokens = tokenizer(example["text"], truncation=False)
    return len(tokens["input_ids"]) <= MAX_SEQ_LENGTH - 10

before = len(dataset)
dataset = dataset.filter(filter_long)
print(f"Filtered: {before} → {len(dataset)} (removed {before - len(dataset)} too-long examples)")

split = dataset.train_test_split(test_size=0.05, seed=42)
train_dataset = split["train"]
eval_dataset = split["test"]

print(f"Train: {len(train_dataset)} | Eval: {len(eval_dataset)}")
```

```python
num_examples = len(train_dataset)

if num_examples < 200:
    epochs = 8
elif num_examples < 500:
    epochs = 6
elif num_examples < 1000:
    epochs = 5
else:
    epochs = 3

print(f"Dataset: {num_examples} examples → {epochs} epochs")

training_args = TrainingArguments(
    output_dir="./code-commenter-output",
    num_train_epochs=epochs,
    per_device_train_batch_size=1,
    gradient_accumulation_steps=8,
    learning_rate=2e-4,
    lr_scheduler_type="cosine",
    warmup_ratio=0.1,
    logging_steps=5,
    eval_strategy="steps",
    eval_steps=25,
    save_strategy="steps",
    save_steps=50,
    save_total_limit=2,
    fp16=not torch.cuda.is_bf16_supported(),
    bf16=torch.cuda.is_bf16_supported(),
    optim="adamw_8bit",
    max_grad_norm=0.3,
    weight_decay=0.01,
    seed=42,
    report_to="none",
)
```

```python
trainer = SFTTrainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
    tokenizer=tokenizer,
    dataset_text_field="text",
    max_seq_length=MAX_SEQ_LENGTH,
    packing=False,
)

steps = num_examples * epochs // 16
print(f"\n{'='*60}")
print(f"  TRAINING WITH UNSLOTH + QWEN 7B")
print(f"  Examples: {num_examples}")
print(f"  Epochs: {epochs}")
print(f"  Effective batch: 4 x 4 = 16")
print(f"  Total steps: ~{steps}")
print(f"{'='*60}\n")

trainer.train()
print("\n✅ Training complete!")
```

```python
SAVE_PATH = "code-commenter-lora"
model.save_pretrained(SAVE_PATH)
tokenizer.save_pretrained(SAVE_PATH)
print(f"Saved to {SAVE_PATH}/")
```

```python
print("\n" + "="*60)
print("TESTING YOUR MODEL")
print("="*60)

FastLanguageModel.for_inference(model)

test_codes = [
    {
        "code": "#include <bits/stdc++.h>\nusing namespace std;\nint main() {\n    int n;\n    cin >> n;\n    vector<int> a(n);\n    for (int i = 0; i < n; i++)\n        cin >> a[i];\n    sort(a.begin(), a.end());\n    cout << a[n-1] - a[0] << endl;\n    return 0;\n}",
        "lang": "C++"
    },
    {
        "code": "n = int(input())\na = list(map(int, input().split()))\na.sort()\nprint(a[-1] - a[0])",
        "lang": "Python"
    },
    {
        "code": "#include <bits/stdc++.h>\nusing namespace std;\nint main() {\n    int t;\n    cin >> t;\n    while (t--) {\n        int l, r, d;\n        cin >> l >> r >> d;\n        if (d < l)\n            cout << d << endl;\n        else\n            cout << ((r / d) + 1) * d << endl;\n    }\n}",
        "lang": "C++ (unseen problem)"
    }
]

for test in test_codes:
    prompt = (
        "<|im_start|>system\n"
        "You are a code commenting assistant. Add short inline comments to the end of each line of code. "
        "Use // for C++ and # for Python. Keep comments under 10 words. "
        "Return ONLY the commented code.<|im_end|>\n"
        "<|im_start|>user\n"
        f"Add inline comments to this code:\n\n{test['code']}<|im_end|>\n"
        "<|im_start|>assistant\n"
    )

    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=512,
            temperature=0.2,
            do_sample=True,
            pad_token_id=tokenizer.eos_token_id,
        )

    result = tokenizer.decode(outputs[0][inputs.input_ids.shape[1]:], skip_special_tokens=True)
    print(f"\n--- {test['lang']} ---")
    print(result.strip())
```
